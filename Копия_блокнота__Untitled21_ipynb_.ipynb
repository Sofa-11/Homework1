```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly as pl
import plotly.express as px
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/TAUforPython/BioMedAI/refs/heads/main/test_datasets/test_data_ECG.csv')
```

```python
df.columns
```

```python
df['ECG_TIME_DATE'] = pd.to_datetime(df['eeg_date '] + ' ' + df['eeg_time '])
```

```python
df['eeg_date '] = pd.to_datetime(df['eeg_date '].astype(str))
df['year'] = df['eeg_date '].dt.year
count_by_year = df.groupby('year')['Count_subj'].sum().reset_index()
```

```python
fig = px.bar(count_by_year, x='year', y='Count_subj', title='Количество проведенных исследований ЭКГ по годам', labels={'Count_subj':'Количество исследований', 'year':'Год'})
fig.show()
```

```python
fig = px.scatter(df,
                 x='t_end',
                 y='t_axis',
                 color='Healthy_Status',
                 title='Распределение классов по параметрам t_end и t_axis',
)
fig.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from keras.models import Sequential
from keras.layers import Dense
```

```python
X = df[['rr_interval', 'p_onset', 'p_end', 'qrs_onset', 'qrs_end', 't_end', 'p_axis', 'qrs_axis', 't_axis']]
Y = df['Healthy_Status']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
model = Sequential()
model.add(Dense(12, input_dim=X_train.shape[1], activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.fit(X_train, Y_train, epochs=150, batch_size=10, verbose=0)
accuracy = model.evaluate(X_test, Y_test)
print(f'Accuracy: {accuracy[1]*100:.2f}%')
```
